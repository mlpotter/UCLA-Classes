```python
import numpy as np
```

```python
d = 5
N = 1000
lambd = 0.05
eta = 100000
w_t = np.random.rand(d,1)
```

```python
1/eta
```

```python
w_true = np.array([[0.5, -.5, 2, 3, 1]]).T
X = np.random.randint(0,10,size=(N,d))
y = np.dot(X,w_true)
```

```python
def soft_threshold(x):
    if x > lambd/eta:
        return x - lambd/eta
    elif abs(x) <= lambd/eta:
        return 0
    elif x < -lambd/eta:
        return x + lambd/eta
        
```

```python
for i in range(1000):
    g = np.matmul(X.T,np.dot(X,w_t)-y)
    w_t = np.array([soft_threshold(wi) for wi in (w_t - 1/eta*g)])
    print(w_t)
```

```python
for i in range(10000):
    g = np.matmul(X.T,np.dot(X,w_t)-y)
    z = w_t - 1/eta * g
    w_t = np.sign(z) * np.maximum(np.abs(z) - lambd/eta,0)
    print(w_t)
```
