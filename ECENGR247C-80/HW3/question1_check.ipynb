```python
import torch
import numpy as np
```

```python
N = 10
M = 5
```

```python
x = torch.rand(N,1)
W = torch.rand((M,N),dtype=torch.float,requires_grad=True)
```

```python
loss = .5*torch.norm(torch.matmul(W.T,torch.matmul(W,x))-x)**2
loss.backward()
```

```python
W_grad_auto = W.grad
```

```python
dg = torch.matmul(torch.matmul(W,x),(torch.matmul(torch.matmul(W.T,W),x)-x).T)
dh = torch.matmul(torch.matmul(W,torch.matmul(torch.matmul(W.T,W),x)-x), x.T)

W_grad_manual = dh+dg
```

```python
(W_grad_manual-W_grad_auto).max()
```

```python
for i in range(10000):
    dg = torch.matmul(torch.matmul(W,x),(torch.matmul(torch.matmul(W.T,W),x)-x).T)
    dh = torch.matmul(torch.matmul(W,torch.matmul(torch.matmul(W.T,W),x)-x), x.T)

    W = W - 0.01 * (dg+dh)
    loss = .5*torch.norm(x-torch.matmul(W.T,torch.matmul(W,x)))**2
    print(loss)
```
